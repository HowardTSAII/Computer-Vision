```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

if __name__ == '__main__':
    anno_file = 'annotation.txt'
    list_of_lists =[]
    bbox = []
    widths = []
    ratio =[]
    rangewid = []
    rangeratio = []
    with open(anno_file, 'r') as fp:
        lines = fp.readlines()
        #print(lines)
    # Enter your code here (25%)
    # You might use the following to access each line
        for line in lines:
            annotation = line.strip().split(' ') 
            annotation = annotation[1:]          
            list_of_lists.append(annotation)
            #list = np.reshape(list_of_lists, (1, -1))
        for a in list_of_lists:
            for b in a:
                if int(a[0]) >= 0 and int(a[1]) >= 0 and int(a[2]) >= int(a[0]) and int(a[3]) >= int(a[1]):
                    bbox.append(b)
        bbox = np.reshape(bbox, (-1, 4))
        #print(bbox)
```

```python
# Step 2b.2
bbox = bbox.tolist()
for i in bbox:
    for j in i:
        a = int(i[2]) - int(i[0]) + 1
    widths.append(a)
print(widths)
for i in widths:
    if i < 10:
        rangewid.append("#1")
    if i >= 10 and i < 20:
        rangewid.append("#2")
    if i >= 20 and i < 30:
        rangewid.append("#3")
    if i >= 30 and i < 40:
        rangewid.append("#4")
    if i >= 40 and i < 50:
        rangewid.append("#5")
    if i >= 50:
        rangewid.append("#6")        
print(rangewid)
```

```python
# Step 2b.3
for i in bbox:
    for j in i:
        a = (int(i[2]) - int(i[0]) + 1) / (int(i[3]) - int(i[1]) + 1)
    ratio.append(round(a ,3))
print(ratio)
for i in ratio:
    if i < 0.6:
        rangeratio.append("#1")
    if i >= 0.6 and i < 0.7:
        rangeratio.append("#2")
    if i >= 0.7 and i < 0.8:
        rangeratio.append("#3")
    if i >= 0.8 and i < 0.9:
        rangeratio.append("#4")
    if i >= 0.9 and i < 1.0:
        rangeratio.append("#5")
    if i >= 1.0:
        rangeratio.append("#6")        
print(rangeratio)
```

```python
# Step 2b.4
widths = np.array(widths)
plt.hist(widths, bins = 10, range = (0, 101), rwidth = 0.85)
plt.xlabel('Widths')
plt.ylabel('Counts')
```
