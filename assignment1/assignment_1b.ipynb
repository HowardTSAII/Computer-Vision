```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
import math
%matplotlib inline

def show_gradient_descent(A, b):
#  enter your code here, and remove the following line (30%)
    x = np.array([[-2], [2]]) # initial condition
    alpha = 0.02 # learning rate
    max_iter = 1000
    f =  (np.linalg.norm(np.dot(A, x) - np.array(b))) ** 2
    np1, np2, np3 =  np.empty(shape=39), np.empty(shape=39), np.empty(shape=39)
    for k in range(max_iter):
        print (k, (round(x[0,0], 4),round(x[1,0], 4)), round(f, 8))
        np1[k] = round(x[0,0], 4)
        np2[k] = round(x[1,0], 4)
        np3[k] = round(f, 8)
        x_prev = x
        𝐀𝑇𝐀x = np.dot(np.dot(A.T, A), x)
        𝐀𝑇𝐛 = np.dot(A.T, b)
        x = np.array(x) - np.array((alpha * (2 * (𝐀𝑇𝐀x - 𝐀𝑇𝐛)))) # 𝛻𝑓(𝑥) = 2(𝐀𝑇𝐀x − 𝐀𝑇𝐛)
        if np.linalg.norm((x - x_prev), 1) <= math.pow(10,-8):
            break
        f = (np.linalg.norm(np.dot(A, x) - np.array(b))) ** 2
    ans = np.vstack([np1,np2,np3])
    return ans
```

```python
def draw_gradient_descent(A, b):
    xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
    F1 = A[0][0]*xx + A[0][1]*yy - b[0];
    F2 = A[1][0]*xx + A[1][1]*yy - b[1];
    zz = F1*F1 + F2*F2;
    plt3d = plt.figure().add_subplot(projection='3d')
    plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm)
    
#  enter your code here (14%)
    ans = (show_gradient_descent(A, b))
    print(ans)
    plt3d.scatter(ans[0], ans[1], ans[2], c='red')
    plt3d.plot(ans[0], ans[1], c='blue')
    plt.show(plt3d)
```
